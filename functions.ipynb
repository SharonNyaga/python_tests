```python
## Write a function that takes a list of numbers and returns their sum
def sum_list(numbers):
    return sum(numbers)

numbers = [5,13,17,23]
result = sum_list(numbers)
print("The sum of the list is:", result)
```

```python
## Create a function that returns the maximum number from a list.
def max_list(numbers2):
    return max(numbers2)

numbers2 = [13, 19,76,78,90]
result = max_list(numbers2)
print("The max number is:", result)
```

```python
sum_1 = 0
def summation(numbers):
    sum_1 = 0
    for number in numbers:
        sum_1 = sum_1 + number
        print("the sum of the list is:", sum_1)


numbers = [5,13,17,23]
result = summation(numbers)
```

```python
##Write a function that takes a dictionary and returns all its keys as a list.
def dict_keys(dictionary):
    return list(dictionary.keys())

dictionary = {
    "name": "Viona",
    "num1": 189,
    "fruits": ["apple", "orange", "pineapple"]
}
result = dict_keys(dictionary)
print("My dict_keys are:", result)


```

```python
## Write a function that accepts a tuple of numbers and returns the average.
def avg_tuple(tuple_1):
    return sum(tuple_1)/len(tuple_1)

tuple_1 = (1,3,5,7)

result = avg_tuple(tuple_1)
print("my tuple is:", result)
```

```python
## Create a function that takes a list and removes all duplicates.
def remove_duplicates(my_list):
    return list(set(my_list))

nums = [2,2,5,7,7,9, 11, 12, 13,13, 13, 17, 23,23, 29,29]

result = remove_duplicates(nums)
print("my clean list is:", result)


```

```python
## Write a function that counts how many times each word appears in a list of strings
def string_count(words):
    counts = {}
    for word in words:
        counts[word] = counts.get(word, 0) +1
    return counts

print(string_count(['carrots','sukuma', 'spinach', 'cabbage', 'carrots', 'spinach']))
```

```python
##Example : Nested if Statements
age = 20
has_ticket = False

if age >= 18:
    if has_ticket:
        print("You can enter the event.")
    else:
        print("You need a ticket.")
else:
    print("You must be 18 or older to enter.")
```

```python
## Example : Multiple Conditions with and, or
age = 12
has_id = True

if age >= 18 and has_id:
    print("Entry allowed.")
else:
    print("Entry denied.")
```
