```python
#Write a program that counts how many times each word appears in a list of words: ["apple", "banana", "apple", "cherry", "banana", "banana"].
words = ["apple", "banana", "apple", "cherry", "banana","banana"]
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

for word, count in word_count.items():
    print(f"{word}: {count}")
```

```python
#using get
words = ["apple", "banana", "apple", "cherry", "banana","banana"]
word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

for word, count in word_count.items():
    print(f"{word}: {count}")

```

```python
# Create a list of dictionaries, e.g., people = [{"name":"Alice","age":30}, {"name":"Bob","age":17}]. Loop through it and print for each person whether they are adult or minor.
people = [
    {"name": "Alice","age": 30},
    {"name": "Bob", "age": 17}
    ]

for person in people:
    name = person["name"]
    age = person["age"]

    if age  >= 18:     ##must be inside the loop so that we get both Alice and Bob evaluated
     print(f"{name} is an Adult")
    else:
     print(f"{name} is a Minor")
```

```python
#Given a list of numbers, ask the user to input the numbers, then compute: number of positive, negative, and zero values using conditionals.
numbers_1 = input("Enter numbers separated by commas: ").split(",")

positive = 0
negative = 0
zero = 0

for num in numbers_1:
    num = int(num.strip())
    if num > 0:
        positive += 1
    elif num < 0:
        negative += 1
    else:
        zero += 1

print("Positive numbers:", positive)
print("Negative numbers:", negative)
print("Zeros:", zero)
```
