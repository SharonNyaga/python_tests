```python
fruits = ["apple", "banana", "cherry", "orange"]
```

```python
## Using a for loop
for fruit in fruits:
    print(fruit)
```

```python
## Using range() and index
for i in range(len(fruits)):
    print(fruits[i])
```

```python
## Using a while loop
i = 0
while i < len(fruits):
    print(fruits[i])
    i += 1
```

```python
#Unpack the tuple ("Tom", 24, "Kenya") into 3 variables and print them.
t = ("Tom", 24, "Kenya")
name, age, country = t
print(name, age, country)
```

```python
#Swap two variables using tuple unpacking (a = 5, b = 10).
a, b = 5, 10
a, b = b, a
print(a, b)
```

```python
#Given coords = (3, 4), compute the distance from origin using unpacking.
coords = (3, 4)
x, y = coords
distance = (x**2 + y**2) ** 0.5
print(distance)
```

```python
#Given t = (1, 2, 3, 4, 5), unpack so that:
#a = 1
#b = [2, 3, 4]
#c = 5
t = (1, 2, 3, 4, 5)
a, *b, c = t
print(a, b, c)
```

```python
#Unpack the tuple (10, 20, 30, 40) into a, *b, c.
a, *b, c = (10, 20, 30, 40)
print(a, b, c)
```

```python
#Unpacking adv methods
#capture the middle
a, *b = [1, 2, 3, 4]
print(a)  # 1
print(b)  # [2, 3, 4]
```

```python
#capture the end
*a, b = [1, 2, 3, 4]
print(a)  # [1, 2, 3]
print(b)  # 4
```

```python
#capture the middle
a, *b, c = [10, 20, 30, 40, 50]
print(a)  # 10
print(c)  # 50
print(b)  # [20, 30, 40]
```
